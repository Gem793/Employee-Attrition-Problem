```python
corr_mat=x.corr().abs()
high_corr_pairs=[] 
cols=corr_mat.columns
for i in range(len(cols)):
    for j in range(len(cols)):
        corr_val=corr_mat.iloc[i,j]
        if corr_val>0.9 and corr_val!=1:
            high_corr_pairs.append((cols[i],cols[j],float(round(corr_val,2))))
high_corr_pairs
```

```python
#Training split and model building 
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression 
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_sm,y_sm,test_size=0.2,random_state=42)
```

```python
cf=RandomForestClassifier()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```

```python
cf=DecisionTreeClassifier()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```

```python
cf=SVC()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```

```python
cf=KNeighborsClassifier()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```

```python
cf=LogisticRegression()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```

```python
cf=XGBClassifier()
md=cf.fit(x_train,y_train)
y_pred=md.predict(x_test)
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)*100
```
